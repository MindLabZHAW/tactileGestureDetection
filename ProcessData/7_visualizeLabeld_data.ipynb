```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
columns_of_interest = ['tau_J0', 'tau_J1', 'tau_J2', 'tau_J3', 'tau_J4', 'tau_J5', 'tau_J6','tau_ext0', 'tau_ext1', 'tau_ext2', 'tau_ext3', 'tau_ext4',
        'tau_ext5', 'tau_ext6','e0', 'e1', 'e2', 'e3', 'e4', 'e5', 'e6','de0', 'de1', 'de2', 'de3', 'de4', 'de5', 'de6','label']
dt_filter = db_ST[columns_of_interest]

features = dt_filter.drop(columns=['label'])
labels = dt_filter['label'].values

pca = PCA(n_components = 2)
reduced_features = pca.fit_transform(features)


# 可视化降维后的结果
plt.figure(figsize=(8, 6))
unique_labels = np.unique(labels)
cluster_colors = ['red', 'blue','green']  # Extend if needed

for i, label in enumerate(unique_labels):
        plt.scatter(reduced_features[labels == label, 0],
                    reduced_features[labels == label, 1], 
                    label=f"Label {label}", 
                    color=cluster_colors[i], 
                    alpha=0.6)   
    
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection of Joint Features')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  # 导入3D绘图工具

# 指定感兴趣的列
columns_of_interest = [
    'tau_J0', 'tau_J1', 'tau_J2', 'tau_J3', 'tau_J4', 'tau_J5', 'tau_J6',
    'tau_ext0', 'tau_ext1', 'tau_ext2', 'tau_ext3', 'tau_ext4', 'tau_ext5', 'tau_ext6',
    'e0', 'e1', 'e2', 'e3', 'e4', 'e5', 'e6',
    'de0', 'de1', 'de2', 'de3', 'de4', 'de5', 'de6', 'label'
]

# 筛选感兴趣的数据
dt_filter = db_ST[columns_of_interest]

# 分离特征和标签
features = dt_filter.drop(columns=['label'])  # 移除标签列
labels = dt_filter['label'].values            # 转换为 NumPy 数组以便布尔索引

# 进行PCA降维，选择3个主成分
pca = PCA(n_components=3)
reduced_features = pca.fit_transform(features)

# 可视化降维后的结果
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')  # 创建3D坐标轴

# 获取唯一标签
unique_labels = np.unique(labels)
cluster_colors = ['red', 'blue', 'green']  # 如果有更多的标签可以扩展

# 使用循环分别绘制不同标签的点
for i, label in enumerate(unique_labels):
    ax.scatter(
        reduced_features[labels == label, 0],  # 第1主成分
        reduced_features[labels == label, 1],  # 第2主成分
        reduced_features[labels == label, 2],  # 第3主成分
        label=f"Label {label}",
        color=cluster_colors[i],
        alpha=0.6
    )

# 设置坐标轴标签
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

# 添加标题和图例
plt.title('3D PCA Projection of Joint Features')
plt.legend()

# 显示图形
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
reduced_features_tsne = tsne.fit_transform(features)

# 绘制t-SNE降维后的结果
plt.figure(figsize=(8, 6))
for i, label in enumerate(unique_labels):
    plt.scatter(
        reduced_features_tsne[labels == label, 0], 
        reduced_features_tsne[labels == label, 1], 
        label=f"Label {label}", 
        color=cluster_colors[i], 
        alpha=0.6
    )

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Projection of Joint Features')
plt.legend()
plt.show()
```

```python
import umap

# 指定感兴趣的列
columns_of_interest = [
    'tau_J0', 'tau_J1', 'tau_J2', 'tau_J3', 'tau_J4', 'tau_J5', 'tau_J6',
    'tau_ext0', 'tau_ext1', 'tau_ext2', 'tau_ext3', 'tau_ext4', 'tau_ext5', 'tau_ext6',
    'e0', 'e1', 'e2', 'e3', 'e4', 'e5', 'e6',
    'de0', 'de1', 'de2', 'de3', 'de4', 'de5', 'de6', 'label'
]

# 筛选感兴趣的数据
dt_filter = db_ST[columns_of_interest]

# 分离特征和标签
features = dt_filter.drop(columns=['label'])  # 移除标签列
labels = dt_filter['label'].values 



# 创建UMAP实例并进行降维
umap_model = umap.UMAP(n_components=2)  # 选择降维到2维
reduced_features_umap = umap_model.fit_transform(features)

# 可视化UMAP降维后的数据
plt.figure(figsize=(8, 6))
unique_labels = np.unique(labels)
cluster_colors = ['red', 'blue', 'green']  # 根据类别数量调整颜色

for i, label in enumerate(unique_labels):
    plt.scatter(reduced_features_umap[labels == label, 0], 
                reduced_features_umap[labels == label, 1], 
                label=f"Label {label}", 
                color=cluster_colors[i], 
                alpha=0.6)

plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.title('UMAP Projection of Joint Features')
plt.legend()
plt.show()
```
