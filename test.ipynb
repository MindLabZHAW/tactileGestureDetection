```python
import json
import pandas as pd
import os
import plotly.express as px
import plotly.io as pio
import cufflinks as cf
from plotly.offline import iplot, init_notebook_mode

# Enable Plotly and Cufflinks offline mode
cf.go_offline(connected=True)
init_notebook_mode(connected=True)
```

```python
data = pd.read_csv("data/0719-7RST-1/tau_J.csv")
data.head()
```

```python
import pandas as pd
import builtins

del len
len = builtins.len

# Convert timestamps to datetime
data["timestamp"] = pd.to_datetime(data["timestamp"])

# Print first timestamp to verify
print("First timestamp:", data["timestamp"][0])

# Calculate the length of the timestamp list
length = len(data["timestamp"])

# Calculate duration
duration = data["timestamp"].iloc[length - 1] - data["timestamp"].iloc[0]

print("Duration:", duration)
```

```python
import plotly.express as px
df = pd.DataFrame(data)
#  0 
df["time_diff"] = (df["timestamp"] - df["timestamp"].min()).dt.total_seconds()

# 
for column in df.columns[1:-1]:  #  timestamp  time_diff 
    fig = px.line(df, x="time_diff", y=column, title=f'Waveform of {column}')
    fig.update_layout(
        xaxis_title="Time (seconds)",
        yaxis_title=column,
        xaxis=dict(tickmode='linear', tick0=0, dtick=5),
        hovermode='x unified',
        width=1000,
        height=600
    )
    fig.show()
```
